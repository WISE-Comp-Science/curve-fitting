```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
f = 10.0 # Hertz
t = np.linspace(-0.5, 0.5, 1000)
x = 10.0*np.sin(2.0*np.pi*f*(t + 0.04)) + 1.0*np.random.randn(len(t))
plt.plot(t, x, marker='.', linestyle='none')
```

```python
f = open('scope_data.txt', 'w')
f.write('#          t                    x\n')
for ti, xi in zip(t, x):
    f.write('{}     {}\n'.format(ti, xi))
f.close()
```
